```python
import pandas as pd

# question 1:
df_erp = pd.read_csv('erp.xlsx - Sheet1.csv', sep=',')
df_lia = pd.read_csv('liaison.xlsx - Sheet1.csv', sep=',')
df_web = pd.read_csv('web.xlsx - Sheet1.csv', sep=',')

df_erp
df_lia
df_web
```

```python
# lier la dataframe erg avec la frame liaison 
a = pd.merge(df_erp, df_lia)
a
```

```python
df_web = df_web.rename(columns = {'sku':'id_web'}, inplace=True)
# modifie directement dans df web grace à inplace
# modifier le nom de l'id du fichier web pour pouvoir lier les tables
```

```python
df_general = pd.merge(a, df_web)
```

```python
df_general
# table complète avec données erp et données web
```

```python
# enlève les élèments de la colone id_web quand il n'y a pas d'id (car ce sont des doublons)
df_general = df_general[df_general["id_web"].notnull()]
# enlève les doublons de post_type car ils représentent des noms de produits doublés
df_general = df_general[df_general["post_type"] == "product"]
```

```python
df_general.to_csv("df_general.csv", sep='\t')
```
